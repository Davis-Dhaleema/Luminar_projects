Naive Bayes Algorithm
* It is a supervised learning alogrirthm based on bayes theorem.
* Simplest and most effect algorithm for fast machine learning models helping quick predictions.
* It is probabilistic classifier ,predicts on the basis of probability.

```python
weather = ['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny',
           'Rainy','Sunny','Overcast','Overcast','Rainy']
temp = ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild']
play = ['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
```

Steps for converting the strings labels into numbers(ie, categorical values)

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
weather_encoded=le.fit_transform(weather)
temp_encoded = le.fit_transform(temp)
play_encoded=le.fit_transform(play)
print(weather_encoded)
print(temp_encoded)
print(play_encoded)
```

Zipping features together, here weather_encoded and temp_encoded

```python
features=zip(weather_encoded,temp_encoded)
features=list(features)
features
```

KNOWING THE TYPE OF NAIVE BAYES ALGORITH TO USE

Types of Naive Bayes Algorithm

1) Gaussian Algorithm : The predictors takes continuous values instead of dicrete.

2) Multinomial : Primarily used for classification problems, ie mainly for categorical values

3) Bernoulli : Works similar to the mutlinomial classifier, but the predictors are independent Boolean variables



```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(features,play_encoded)

```

Prediction

```python
predicted=model.predict([[2,1]])
print('predicted',predicted)
```
