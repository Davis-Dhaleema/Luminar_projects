LUNG CANCER PREDICTION

```python
import pandas as pd
df=pd.read_csv('/content/lung_cancer_examples.csv')
df
```

```python
x=df.iloc[:,2:-1]
y=df.iloc[:,-1]
```

```python
from collections import Counter
print('the number of classes',Counter(y))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)
print(y_pred_knn)
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay
acc_knn=accuracy_score(y_test,y_pred_knn)
print('Accuracy using KNN : ',acc_knn)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred_knn)
```

```python
from sklearn.naive_bayes import GaussianNB
gaus=GaussianNB()
gaus.fit(x_train,y_train)
y_pred_NB=gaus.predict(x_test)
print(y_pred_NB)
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay
acc_NB=accuracy_score(y_test,y_pred_NB)
print('Accuracy using NB : ',acc_NB)
```

```python
from sklearn.svm import SVC
svc_model=SVC()
svc_model.fit(x_train,y_train)
y_pred_svc=svc_model.predict(x_test)
print(y_pred_svc)
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay
acc_svc=accuracy_score(y_test,y_pred_svc)
print('Accuracy using SVC : ',acc_svc)
```
