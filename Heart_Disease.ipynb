 ***Heart Disease Dataset***

Import data

```python
import pandas as pd
data=pd.read_csv('/content/heart.csv')
data
```

Divide into input and target

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
x
```

Divide into train and test dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=5)
x_test
```

Counting the target classes

```python
from collections import Counter
print('The number classes',Counter(y))
```

Describing the dataset

```python
data.describe()
```

Standarizing the larger values

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_test
```

Prediction modelling

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

Performance checking

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay,confusion_matrix
acc=accuracy_score(y_test,y_pred)
print('Accuracy Score = ',acc)
conf=confusion_matrix(y_test,y_pred)
print('Confusion Matrix\n',conf)
labels=['Heart patient','Not Heart patient']
lab=ConfusionMatrixDisplay(conf,display_labels=labels)
lab.plot()
```

Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))            #support is the number of instances
```
