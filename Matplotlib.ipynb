# **matplotlib is a library used for plotting the data and visualizing them.**

# **pyplot in matplotlib is used for displaying features**

## ***1) Plotting point (x,y)***

```python
import matplotlib.pyplot as plt
x=[6,9,6,2,12,7]
y=[3,5,8,9,1,5]
plt.plot(x,y,'*')
plt.show()
```

## ***2) One variable***

```python
x=[13,45,38,9,1,50]
plt.plot(x,'*:b')        # here * -- point and :-- line  , b -- colour
plt.show()
```

## ***3) Two variable***

```python
x=[6,9,6,2,12,7]
y=[3,5,8,9,1,5]
plt.plot(x,y)
plt.show()
```

## ***linestyle and color***

```python
plt.plot(x,linestyle='dashed',color='r')
```

## ***Linestyle, color and linewidth***

```python
plt.plot(x,linestyle='dotted',color='g',linewidth='5')
```

# ***Two lines in one graph***

```python
a=[1,2,3,4]
b=[1,6,7,8]
c=[4,2,8,1]
d=[1,7,6,5]
plt.plot(a,c,linewidth='2',color='red')
plt.plot(b,d,linewidth='2',color='green')
plt.show()
```

## ***X and Y labelling and heading labelling and grid***

```python
x=[150,156,157]
y=[45,65,50]
plt.plot(x,y,color='g',linestyle='dashed')
plt.xlabel('height in cm')
plt.ylabel('weight in kg')
plt.title('BMI',loc='left')
plt.grid()
```


## ***y grid ,color, thickness***


```python
x=[150,156,157]
y=[45,65,50]
plt.plot(x,y,color='g',linestyle='dashed')
plt.xlabel('height in cm')
plt.ylabel('weight in kg')
plt.title('BMI',loc='left')
plt.grid(axis='y',linestyle='dashed',color='b')
```

## **x grid,color,thickness,style**

```python
x=[150,156,157]
y=[45,65,50]
plt.plot(x,y,color='g',linestyle='dashed')
plt.xlabel('height in cm')
plt.ylabel('weight in kg')
plt.title('BMI',loc='left')
plt.grid(axis='x',color='r',linestyle='dotted',linewidth='2')
```

# ***subplot***

Subplot function is used for mutliple plots.

```python
a=[1,2,3,4]
b=[1,6,7,8]
plt.subplot(1,2,2)   #1-row,2-number of column,2-postion of graph
plt.plot(a,b,linestyle='dashed')
plt.title('Graph 1')
c=[4,2,8,1]
d=[1,7,6,5]
plt.subplot(1,2,1)
plt.plot(d,c,linestyle='dotted')
plt.title('Graph 2')
plt.suptitle('Combo')
```

## ***Scatterplot***


```python
x=[150,256,357,152,149,162]
y=[245,265,350,676,165,263]
plt.scatter(x,y)
a=[100,200,300,400]
b=[150,106,367,248]
plt.scatter(a,b)
```

## ***Bar chart***

```python
a=[10,15,20,25]
b=[15,16,36,24]
plt.bar(a,b)
```

# ***Horizontal Bar chart***

```python
a=[10,15,20,25]
b=[15,16,36,24]
plt.barh(a,b,color='y')
```

## ***Histogram***

```python
x=[12,20,30,40,12,12,12,12,30,30,30,40]
plt.hist(x,color='red')
```

## ***Pie chart***

```python
y=[10,20,30,40,50]
mylabels=['apple','orange','plum','strawberry','blueberry']
plt.pie(y,labels=mylabels)
plt.legend()
```

```python
a=[7,6,5,9,3,4]
b=[3,2,5,8,9,7]
c=[7,4,3,8,6,5]
d=[8,7,9,4,5,6]
plt.subplot(1,4,3)
plt.plot(a,b,color='r')
plt.title('Graph 1')
plt.subplot(1,4,1)
plt.plot(b,d,color='y')
plt.title('Graph 2')
plt.subplot(1,4,4)
plt.plot(a,c,color='b')
plt.title('Graph 3')
plt.subplot(1,4,2)
plt.plot(a,d,color='g')
plt.title('Graph 4')
plt.suptitle('GRAPHS')
```
